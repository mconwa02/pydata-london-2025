## Grid world demo
 - Simple demonstration of the classical grid-world RL environment
 - This notebook demonstrates the following:
      - A random agent (very useless!)
      - A Q-learning agent ( much more intelligent and useful!)

#### Imports

```python
from dataclasses import dataclass, asdict
import sys, os
from pprint import pprint
import pandas as pd
from IPython.display import display

sys.path.insert(1, "../")
from src.main.rl_agents.random_agent import RandomAgent
from src.main.rl_agents.q_learning_agent import QLearningAgent
from src.main.evaluations.evaluate_rl_agent import EvaluateRLAgent
from src.main.configs.random_agent_config import RandomAgentConfig
from src.main.configs.q_learning_agent_configs import QLearningAgentConfig
import src.main.configs.global_configs as configs
from src.main.utility.utils import Helpers
from src.main.utility.chart_results import ChartResults
```

#### Demostrate the "Random" agent

```python
def demoRandomAgentShowGrid(is_show_grid: bool = True):
    """
    Demos random agent navigation in the grid
    """
    random_agent_configs = RandomAgentConfig()
    if is_show_grid:
        random_agent_configs.render_mode = "human"
    else:
        random_agent_configs.render_mode = "rgb_array"
    configs = asdict(random_agent_configs)
    random_agent = RandomAgent(**configs)
    random_agent.run()
    random_agent.plotRewardCurves()

```

#### Random agent with Grid visualization

```python
demoRandomAgentShowGrid(is_show_grid=True)
```

#### Random agent no Grid visualization

```python
demoRandomAgentShowGrid(is_show_grid=False)
```

#### Demostrate a usefull agent using Q-learning RL

```python
def demoQLearningAgentTraining():
    """
    Demos Q-learning agent training as it navigates in the grid
    """
    agent_configs = asdict(QLearningAgentConfig())
    q_learning_agent = QLearningAgent(**agent_configs)
    _, cum_reward, _ = q_learning_agent.run()
    Helpers.serialObject(q_learning_agent, configs.GRIDWORLD_QL_MODEL_FILE_PATH)
    ChartResults.plotRewardCurve(cum_reward, window_size=200)
```

```python
demoQLearningAgentTraining()
```

```python
def demoQLearningAgentEvaluation():
    """
    Demos Q-learning agent evaluation/testing as it navigates in the grid
    """
    agent = Helpers.deserializeObject(configs.GRIDWORLD_QL_MODEL_FILE_PATH)
    rl_evaluator = EvaluateRLAgent(agent=agent)
    print(f"Trained Q-table:\n")
    q_table_df = pd.DataFrame(agent.q_table)
    display(q_table_df)
    print("\n\n")
    rl_evaluator.evaluate()
```

```python
demoQLearningAgentEvaluation()
```
